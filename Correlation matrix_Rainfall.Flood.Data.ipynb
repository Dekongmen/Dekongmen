```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('D:/Upper West Flood Risks_Manuscript/Datasets/NADMO UWR/Correlation/RainFloodData.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
corr=df.corr()
corr
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(corr,square=True, cmap='crest', linewidths=1, annot=True, vmin=-1, vmax=1)
plt.show()
```

```python
matrix=np.triu(corr)
matrix
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(corr, mask=matrix, square=True, cmap='bwr',  linewidths=0.5, vmax=1, vmin=-1, annot=True)
plt.show()
```
