```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('/content/Flood disaster_Data - Origin.csv')
```

```python
df.head()
```

```python
corr=df.corr()
corr
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(corr,square=True, cmap='bwr', vmin=-1, vmax=1, linewidths=0.5)
plt.show()
```

```python
matrix=np.triu(corr)
matrix
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(corr, mask=matrix, square=True, cmap='bwr',  linewidths=0.5, annot=True)
plt.show()
```
